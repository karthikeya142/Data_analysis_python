```python
#526. Merging Datasets
#we need to understand that most of the data analysis operations that are usually performed
#on a data set involve data preparation, loading, cleaning and transformation.
#But even before analyzing the data, there's a whole bunch of operations which you need to perform before analyzing the data.
#combining two data sets together is one such kind of an operation as well.Now the question is why exactly would you combine a different data sets together?
# let's suppose you want to analyze data which comes from an external system.Or it comes from an external source.
#So not all the data is going to be included in one database or one file.
#So in such cases when the data is split across multiple files, you need to go ahead and combine that data together.
#And this combining operations of two or more data sets together is called as merging two data sets together.
#it's essential to learn how exactly we could merge multiple data sets.
```

```python
from pandas import DataFrame
df1 =DataFrame({'Fruit':['Apples','orange','banana','peach'],'prices 2023':[10,20,30,40]})
df2 =DataFrame({'Fruit':['Apples','orange','mango'],'prices 2024':[12,33,36]})
df1
```

```python
df2
```

```python
import pandas as pd
pd.merge(df1,df2)
```

```python
#use outer join
pd.merge(df1,df2, how ="outer")
```

```python
#use left join
#you want to have all the elements of the first data frame, but you do not want to have the elements of the second data frame.
pd.merge(df1,df2, how ="left")
```

```python
#use right join
#it will take all the elements which are present in the second data frame, and only those elements from the first data frame which match up with the second data frame.
pd.merge(df1,df2, how ="right")
```

```python
#535. Frequencies In Time Series   (Freq: D,)
#In this lecture, let's learn about frequencies in a time series.
# now the question is  how exactly can we change the frequency from date to something else.
pd.date_range(start='1/1/2024',periods=20)
```

```python
#dtype='datetime64[ns]', freq='D')
#So here as the frequency is set to days, the 20 over here is going to be counted as 20 days, and this will be reset after every day has passed.
#Now what if you actually want to go ahead and change this particular value after every single hour, instead of changing it after every single day?
#So in that particular case, what we have to do is we have to define the frequency for this as well.
#So the by default frequency parameter is set to capital D which means days.
#Now if you want to set it to something like hours you could say comma and then type in freq and set this parameter to h which stands for hours.
pd.date_range(start='1/1/2024',periods=20,freq='h')
#So if you run this right now here, you're only getting one single date here, which is 1st January
```

```python
#Across all the values which you have now, the thing which changes here is actually the time, and the time actually changes by the hour
# (dtype='datetime64[ns]', freq='h')

#Now here, as we have not specified any time, the time actually started at this particular value right here But you could go ahead and define your own time as well.
#So for example, if you say, all right, you don't want to start the clock at zero hours, zero minutes zero seconds
#So if you do this now, the clock will instead start at this time right here.
#So now let's suppose instead of using the frequency to hours let's say you want to keep the frequency So let me go ahead and add days here.

pd.date_range(start='1/1/2024 02:00:00',periods=20,freq='d')
```

```python
#So this is obviously going to go ahead and change the date.But you'll be able to see that the time over here remains constant.
#Now let's suppose you want to set the frequency to business days instead of using just days.
#So let's suppose you are creating an e-commerce application, and you want to estimate the delivery date for a particular order.
#And in that particular case, instead of counting days, you might want to count business days. So let's see what happens if we do that.
#So in order to use a business days as frequency, replace D with B.

pd.date_range(start='1/1/2024 02:00:00',periods=20,freq='b')
```

```python
#And if you run this code, as you can see we have one, then 234, five and then six and seven are missing
#So this frequency B is only going to count the business days.
#And it's actually going to eliminate all the non business days like weekends.Now there's one more thing which you could do here.
#So for example if I go ahead and set the frequency to H here.As you can see the value changes every single hour.
pd.date_range(start='1/1/2024 02:00:00',periods=20,freq='h')
```

```python
#But what if you want to change the value for every nine hours.
#So in that case you simply have to replace H with nine hours.
#And that's actually going to go ahead and modify everything for you.
pd.date_range(start='1/1/2024 02:00:00',periods=20,freq='9h')
```

```python
#So if you mention nine here the first value which we have here is two hours because we had specified the initial value here.
#But if you take a look at the second value here, what it does is that it basically proceeds ahead nine hours in time And then it's going to create a second value here, which is at time 11.
#

#So let's suppose you want to go ahead and create a value after every nine hours and 34 minutes.
#So let's suppose you want to go ahead and create a value after every nine hours and 34 minutes.
#And if you run this now as you can see the first value is at two.
pd.date_range(start='1/1/2024 02:00:00',periods=20,freq='9h34min')
```

```python
#Now here we have actually passed in some aliases.For example, for hours we have H.
#For business days we have B, then for nine hours we have a multiple of nine over here as well.
#And if you want to get a list of all of these frequency aliases, you could go to the pandas official
#documentation and search for pandas underscore date range function.
#
pd.date_range(start='1/1/2024 02:00:00',periods=20,freq='3ME',tz='Asia/Kolkata')
```

```python
pd.date_range(start="2001-06-03", periods=10, freq="99YS", unit="s")
```

```python
#536. Shifting Time Series
#in this lecture, let's learn about shifting data in a time series.
#So the data which you might get in a time series might sometimes not be perfectly aligned.
#So for example, it might be lagging behind or it might be way ahead of the time.
#So in such cases when data is not aligned properly and you need to readjust your data backwards or forwards,in that particular case you need to go ahead and shift the data.
#So there's a special method called as shift which allows us to go ahead and shift a particular data either backwards or forwards in time.
from pandas import Series
import numpy as np
ts =Series(np.random.randn(10),index=pd.date_range('1/1/2024',periods=10,freq='ME'))
ts
```

```python
#Now here let's suppose this data which you are getting here is let's suppose it's actually seven months behind
#So in that particular case what we have to do is we have to shift this data seven months ahead.
#So by shifting this data seven months ahead we basically mean that this value which we have up over here, this should actually be present at the seventh month.
#So in order to do this what we do is we say the time series dot shift.And over here I simply need to pass in the value by which I want to shift the data.
#So I want to shift the data by seven months here.
ts.shift(7)
```

```python
#And if I run this, as you can see the data is now shifted by seven months.
ts.shift(7)
#However, the value which we had at the first position, it actually got placed at the eighth position,and the rest of the values follow from there.
#
```

```python
#Now let's suppose here you do not want to shift the data by months, but let's say the data which you have is actually off by a couple of days.
#So if I go ahead and run this, let's suppose these are the specific dates on which employees of your company had to be paid salaries.
#But let's suppose some event happens And let's suppose everything gets delayed by 20 days.
#So imagine for a second that you want to delay the salaries for each month by 20 days over here.
#So in order to do that here you say TZ dot shift.
#But right now if I do 20, what would happen is it will actually go ahead and shift the data by 20 months not 20 days
#All right, I want to use the frequency as days not months.
ts.shift(20,freq='d')
#So now if I go ahead and run this all the data which you have up over here, as you can see it is now
```

```python
#Now, another thing which you could do here is that you could also shift the data in a backward direction
#But we could also shift the data backward as well.So let's suppose here I want to shift the data ten days before.
#So here instead of saying ten I would now say minus ten.
ts.shift(-10,freq='d')
```

```python
#So this is how you could go ahead and make use of the shift operations in order to either shift your data forward or backward in time.
```
